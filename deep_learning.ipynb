## **Neuron**

* A **neuron** is the **basic unit** of a neural network, just like a **brain cell** in the human body.
* It **receives input, processes it, and passes the output** to the next neuron.

neuron.png

## **Neural Network**

* A Neural Network is a machine learning model inspired by the human brain. It consists of layers of neurons that process data and learn patterns.

* **Takes inputs ➝ Applies weights & activation ➝ Produces output**

* Uses backpropagation to improve accuracy

**Types**

* ANN (Artificial Neural Network) – Basic neural network

* CNN (Convolutional Neural Network) – Used for image processing

* RNN (Recurrent Neural Network) – Used for sequential data (text, speech)

deepLearn_2_EN.png

### **Activation Function**
An **activation function** decides whether a neuron should be **activated (ON)** or **not (OFF)**. It **adds non-linearity** so the neural network can learn complex patterns.

Types of Activation Functions:
* **ReLU** (Rectified Linear Unit) – Used in hidden layers, fast & efficient
* **Sigmoid** – Used for probability-based outputs (0 to 1)
* **Tanh** – Outputs between -1 to 1, better than sigmoid
* **Softmax** – Used for multi-class classification

**Without activation, a neural network would behave like a simple linear model.**

###**Backpropagation**
**Backpropagation** (Backward Propagation of Errors) is the **learning process** in a neural network, where errors are sent backward through the layers to adjust weights. It minimizes errors using **Gradient Descent** and repeats until the network learns correctly.

**Backpropagation is what makes a neural network learn and improve.**

## **Deep Learning**

**Deep Learning** is a branch of Machine Learning that uses **Neural Networks** with multiple hidden layers to **learn patterns** from data.

* **ML –** Works well for structured data, needs manual feature selection
* **DL –** Works with unstructured data (images, text, audio), learns features automatically

## **Artificial Neural Networks (ANN) → Basic Model**

* ANN is a **fully connected** network best suited for structured data tasks like classification and regression.

   * **Radial Basis Function Network (RBFN)** → Uses distance-based activation

### **1. Feedforward Neural Network (FNN)** → Simple multi-layer perceptron

A **FNN** is the simplest type of Artificial Neural Network (ANN) where data flows **strictly in one direction** (from input → hidden layers → output), without any loops or feedback connections.

* Take input data.  
* Pass through layers (Forward Propagation).  
* Compare prediction with actual value (Loss Calculation).  
* Adjust weights to reduce error (Backpropagation).


feedforward-neural-network.png

**Neural networks process data layer by layer, where all neurons in a layer compute simultaneously before passing outputs to the next layer.**

**1. Forward Propagation** (Making Predictions)
  * Each neuron gets inputs, multiplies them by weights, adds a bias, and applies an activation function to get an output.
  
  `Formula for a single neuron:`

    **`𝑦 = Activation ( ∑ ( 𝑤 𝑖 ⋅ 𝑥 𝑖 ) + 𝑏 )`**

  `Where:  𝑥  ​= Inputs,  𝑤  ​= Weights,   𝑏 = Bias`

* The result will be passed to the next layer until the output layer gives a final prediction.

  **Activation Function = Decides whether the neuron should "fire"**


**2. Compute Error** (Loss Function)
  * Compares the prediction to the actual value using a loss function (e.g., Mean Squared Error for regression, Cross-Entropy for classification).

**3. Backpropagation** (Learning from Mistakes)
  * Adjusts weights to reduce the error using Gradient Descent.
Finds the direction in which error decreases the fastest and updates weights accordingly.

  `Gradient Descent Formula:`

    **`𝑤 = 𝑤 − 𝛼 ⋅ ∂ Loss / ∂ 𝑤 ​`**

  `Where:  𝛼 = Learning rate (small step size to update weights)`

### **2. Radial Basis Function Network (RBFN)** → Use distance-based activation

* **RBFN** is a type of ANN that classifies data based on **distance** from **center points**. Instead of just using weights like a normal neural network.


* The network calculates the distance between an input **𝑥**  and the center **𝑐**  of each hidden neuron.

  * If an input is **close** to a center, it gets a **high activation** (strong influence).
  * If it's **far**, it gets a **low activation** (weak influence).
  * It uses **radial basis functions (RBFs) as activation functions** to measure closeness.

rbf.png

**1. Compute Distance from Center Points**
* Each hidden neuron has a center point (selected from training data).
* The neuron calculates the distance between the input and the center point using the Euclidean distance formula:

  **`𝑑 = ∑ ( 𝑥𝑖 − 𝑐𝑖 ) 2`**
  
  Where:  𝑥𝑖  ​= Input values, 𝑐𝑖  ​= Center values

**2. Apply Radial Basis Function** (Gaussian Function)
* A Gaussian function converts distance into an activation value:

    **`𝜙 ( 𝑥 ) = 𝑒^ − ( ||𝑥 − 𝑐 ||^2 / 2 𝜎^2  )`**

  Where:

  𝑐  = Center point

  𝜎  = Spread (determines how wide the function is)

  **∣∣x−c∣∣** is the **Euclidean distance** between input and center.

  * If input is close to center → High activation
  * If input is far from center → Low activation

**3. Compute Output Using Weights**
* The outputs from the hidden layer are multiplied by weights and summed to produce the final output:

  **`𝑦 = ∑ 𝑤𝑖 ⋅ 𝜙 ( 𝑥 )`**

  Where:

  𝑤 𝑖 ​= Weights for each neuron

  𝜙 ( 𝑥 ) = Activation values from hidden neurons

### **Working Steps of FNN & RBFN**  

**1. Take Input Data**  
- **FNN** →  Example: Predict house price based on **size, location, and number of rooms**.  
- **RBFN** →  Example: Detect fraud based on **transaction amount, location, and time**.  


**2. Process Input in Hidden Layers**  
- **FNN** →  
  - **Multiply inputs with weights, add bias**.
    - **`Z = W⋅X + B`**
  - Apply an **activation function** (ReLU, Sigmoid, etc.).  
    - **`A = activation(Z)`**
- **RBFN** →  
  - **Measure distance** between input and predefined RBF centers (found using K-Means clustering).
    - **`Distance = ∣∣X−C∣∣`**  
  - Convert distance into a **similarity score** using a **Gaussian function**.
    - **`ϕ(X) = exp (− (∣∣X−C∣∣^2) / (2.σ^2) )`**


**3. Compute Final Output**  
- **FNN** →   Last layer applies activation (e.g., Softmax) and **generates a prediction** (house price).
    - **`Y = Wout ⋅ A`**

- **RBFN** →   Multiply similarity scores with weights and **compute final result** (fraud or not fraud).
    - **`Y = Wout ⋅ ϕ(X)`**  


**4. Compare with Actual Output (Loss Calculation)**  
- **Both** → Check how far the prediction is from the true value using a **loss function**.  


**5. Adjust Parameters to Improve Accuracy**  
- **FNN (Backpropagation)** →  
  - Compute **error gradients** and **adjust weights** using **gradient descent**.  

- **RBFN (Weight Update)** →  
  - **Update RBF centers using clustering (K-Means)**.  
  - **Update final layer weights using least squares or gradient descent**.  

**6. Repeat Until Convergence**  
- Train multiple **epochs** until the model **makes accurate predictions**.

* **FNN** → Learns by updating **weights (W) at every layer** using backpropagation.  

* **RBFN** → Learns by updating RBF centers (C) using clustering and adjusting final weights (W) using least squares or gradient descent.

## **Convolutional Neural Networks (CNN) → Image Processing**

### **Concept**
**CNN** is a type of neural network designed for **image processing** and **pattern recognition**, CNN **extracts features** from images automatically using **convolution** operations.


* Input Layer – Takes an image (e.g., 28x28 pixels for digit recognition).
* Convolutional Layer – Extracts features (edges, textures, shapes).
* ReLU Activation (Non-Linearity)
* Pooling Layer – Reduces size while keeping important features.
* Fully Connected Layer – Makes the final prediction (e.g., "cat" or "dog").


**Convolution → ReLU → Pooling → Fully Connected Layer**


*All CNNs follow the same basic structure but differ in how they extract features, reduce size, and train deep layers.*


cnn.png

### **CNN steps**

**1. Input Processing**  

- Convert image to a pixel matrix.  
- Resize to a fixed shape.  
- Normalize pixel values for better training.  

  **Architecture-Specific Differences:**  
  * **LeNet** → Small grayscale images (e.g., **32×32**).  
  * **AlexNet** → RGB images (**224×224**), uses normalization.  
  * **VGG** → Uses fixed **224×224** size, standard preprocessing.  
  * **ResNet** → Same as VGG but deeper, adds **batch normalization**.  
  * **Inception** → Uses **multiple input sizes** for different filters.  
  * **MobileNet** → Optimized for mobile, lightweight preprocessing.  
  * **EfficientNet** → Uses **scaling** (changes width, depth, resolution).  


**2. Feature Extraction (Convolution Layers)**

- Apply filters (**kernels**) to detect **edges, textures, objects**.  
- Create **feature maps** from learned patterns.  

  **Architecture-Specific Differences:**  
  - **LeNet** → Uses **5×5 filters** with fewer layers.  
  - **AlexNet** → Uses **11×11 filters**, deeper than LeNet.  
  - **VGG** → Uses multiple **3×3 filters** instead of large ones.  
  - **ResNet** → Adds **skip connections** to prevent vanishing gradients.  
  - **Inception** → Uses **1×1, 3×3, and 5×5 filters at the same time**.  
  - **MobileNet** → Uses **Depthwise Separable Convolutions** to reduce computation.  
  - **EfficientNet** → Uses **MBConv (inverted residual blocks)** for efficiency.  


**3. Activation Function (ReLU/GELU)**

- Replace negative values with zero (**ReLU**) or smooth approximation (**GELU**).  

  **Architecture-Specific Differences:**  
  - **LeNet** → Uses **Sigmoid/Tanh** (old architectures).  
  - **All Modern CNNs** → Use **ReLU** for non-linearity.  
  - **Some Transformers (like Swin)** → Use **GELU** instead of ReLU.  


**4. Pooling (Downsampling)**

- Reduce **feature map size** while keeping key features.  

  **Architecture-Specific Differences:**  
  - **LeNet** → Uses **Average Pooling**.  
  - **AlexNet** → Uses **Max Pooling** to reduce size.  
  - **VGG** → Uses **Max Pooling after every few layers**.  
  - **ResNet** → Same as VGG, but optimized with **batch norm**.  
  - **Inception** → Uses **Max & Average Pooling together**.  
  - **MobileNet & EfficientNet** → Use **Global Average Pooling** instead of FC layers.  


**5. Fully Connected Layers (Dense Layers)**  

- Flatten **feature maps** into a 1D vector.  
- Pass through **fully connected layers**.  
- Use **Softmax activation** for classification.  

  **Architecture-Specific Differences:**  
  - **LeNet** → Fewer neurons, simpler layers.  
  - **AlexNet** → **4096 neurons**, heavy computation.  
  - **VGG** → Fully connected with **two 4096-neuron layers**.  
  - **ResNet** → Similar to VGG but optimized with **fewer parameters**.  
  - **Inception** → Uses **1×1 convolutions** to reduce parameters before FC layers.  
  - **MobileNet & EfficientNet** → **Remove FC layers**, replace with **Global Average Pooling**.  


**6. Loss Calculation & Training (Backpropagation & Optimization)**  

- Compute error using **Loss Function** (e.g., **Cross-Entropy**).  
- Adjust weights using **Backpropagation + Optimizer** (**SGD, Adam, RMSprop**).  
- Train for **multiple epochs** until accuracy improves.  

  **Architecture-Specific Differences:**  
  - **All CNNs** → Use **Cross-Entropy** for classification tasks.  
  - **ResNet & Inception** → Use advanced optimization (**Batch Norm, Adam**).  
  - **MobileNet & EfficientNet** → Optimized for **faster training** on low-power devices.  
  - **Modern CNNs** → Use **Learning Rate Scheduling** & **Dropout** to prevent overfitting.  

  

- **All CNNs** follow the same core steps, but differ in **filter sizes, layer depth, pooling methods, and optimization tricks**.  
- **LeNet** → Basic, **AlexNet** → Deeper, **VGG** → Small filters, **ResNet** → Skip connections, **Inception** → Multi-scale filters, **MobileNet/EfficientNet** → Lightweight.  
- **EfficientNet** is **newer than MobileNet** and uses **scaling tricks** to improve performance **without increasing computation**.  

### **Comparison with FNN & RBFN**

- **FNN vs. RBFN** → Completely different approach (weighted sum vs. radial basis functions).  
- **CNN Architectures** → **Same core method** (Convolution → Activation → Pooling → Fully Connected), but each model **tweaks some parts** to improve performance.  

**CNN Architectures**   
- Uses **Convolution** to extract features  
- Uses **ReLU** for activation  
- Uses **Pooling** for downsampling  
- Uses **Fully Connected Layers** for classification  

**Difference**  
- **Filter sizes** (VGG = small 3×3, AlexNet = large 11×11)  
- **Number of layers** (ResNet = very deep with skip connections)  
- **Pooling methods** (LeNet = Avg Pooling, AlexNet/VGG = Max Pooling)  
- **Efficiency tricks** (MobileNet = Depthwise Separable Conv, Inception = Multiple filter sizes in parallel)  

💡 **Conclusion:**  
- **Unlike FNN & RBFN, CNN architectures do NOT use entirely different approaches.**  
- **Learn one CNN well, and you can easily understand others by noting their small improvements.** 🚀

## **Steps for All RNN Architectures**  

All **Recurrent Neural Networks (RNNs)** follow the **same fundamental process**, but different architectures tweak certain steps to improve memory retention, training stability, and performance.  

---

### **1️⃣ Input Processing**  
- Convert input (e.g., text, time-series) into **sequential data**.  
- Use **word embeddings (for NLP)** or **normalized numerical sequences**.  
- Process one **time step at a time**.  

✔ **Basic RNN → Directly processes sequential input.**  
✔ **LSTM → Stores long-term dependencies better.**  
✔ **GRU → Similar to LSTM but with fewer parameters.**  
✔ **BiLSTM/BiGRU → Processes input **both forward & backward** for better context.  

---

### **2️⃣ Hidden State Update (Recurrent Computation)**  
- Process current input **AND** previous hidden state:  

  \[
  h_t = \text{activation}(W \cdot x_t + U \cdot h_{t-1} + b)
  \]  

- Updates **hidden state** at each time step.  

✔ **Basic RNN → Uses simple recurrence, struggles with long-term dependencies.**  
✔ **LSTM → Uses gates to selectively keep/forget information.**  
✔ **GRU → Uses fewer gates but achieves similar results.**  

---

### **3️⃣ Activation Function (Non-Linearity in Processing)**  
- Apply activation functions to introduce non-linearity.  
- **ReLU** or **Tanh** is used for hidden state updates.  

✔ **Basic RNN → Uses Tanh, but vanishing gradient occurs.**  
✔ **LSTM/GRU → Uses Sigmoid/Tanh inside gates to control memory retention.**  

---

### **4️⃣ Memory Management (Handling Long-Term Dependencies)**  
- Basic RNNs **forget past information** due to **vanishing gradients**.  
- LSTMs & GRUs solve this by using **memory cells & gates**.  

✔ **LSTM → Uses Forget, Input, and Output gates.**  
✔ **GRU → Uses Reset & Update gates (simpler than LSTM).**  

**LSTM Memory Cell Equations:**  
- **Forget Gate:**  
  \[
  f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)
  \]  
- **Input Gate:**  
  \[
  i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)
  \]  
- **Cell State Update:**  
  \[
  C_t = f_t * C_{t-1} + i_t * \tilde{C}_t
  \]  
- **Output Gate:**  
  \[
  o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)
  \]  
- **Hidden State Update:**  
  \[
  h_t = o_t * \tanh(C_t)
  \]  

---

### **5️⃣ Output Computation (Final Predictions)**  
- The last hidden state is passed through **fully connected layers**.  
- Softmax is used for classification tasks.  

✔ **Basic RNN → Outputs at each time step.**  
✔ **LSTM/GRU → Can output at each step or final step.**  
✔ **Seq2Seq (used in NLP) → Uses special "decoder" RNN.**  

---

### **6️⃣ Loss Calculation & Training (Backpropagation Through Time - BPTT)**  
- Compute **loss** (e.g., Cross-Entropy for classification, MSE for regression).  
- Apply **Backpropagation Through Time (BPTT)** to update weights.  
- Use optimizers like **SGD or Adam**.  

✔ **Basic RNN → Faces vanishing gradients.**  
✔ **LSTM/GRU → More stable training due to gating mechanisms.**  

---

### **7️⃣ Repeat Training Until Convergence**  
- Iterate multiple times (epochs) until model performance stabilizes.  

✔ **BiLSTM/BiGRU → Uses forward + backward passes for better context.**  
✔ **Transformers (Next-gen RNN Alternative) → Uses self-attention instead of recurrence.**  

---

### 🔥 **Key Takeaways**  
✔ **All RNNs** follow the same steps but differ in **how they handle memory**.  
✔ **LSTM & GRU** improve long-term dependency handling.  
✔ **BiLSTM/BiGRU** capture **both past & future** dependencies.  
✔ **Transformers (e.g., BERT, GPT) are replacing RNNs** for NLP due to better parallelization.
